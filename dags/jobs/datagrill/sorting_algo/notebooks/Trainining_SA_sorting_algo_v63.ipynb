```python
#import libraries
import warnings
import shap
warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd
import random
import pickle
from google.cloud import bigquery
import math
import re
import json
import xgboost as xgb
from sklearn.metrics import ndcg_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn import preprocessing
from itertools import combinations 
import time
from datetime import datetime
import matplotlib.pyplot as plt 
import seaborn as sns
from google.colab import auth
from google.cloud import bigquery
from google.colab import data_table
import numpy as np
#from comet_ml import Experiment
pd.options.mode.chained_assignment = None


project = 'namshi-datahut' 
location = 'US' 
client = bigquery.Client(project=project, location=location)
data_table.enable_dataframe_formatter()
auth.authenticate_user()
from google.colab import drive
drive.mount('/content/drive')

#----USER INPUTS---#

#Bucket location
buc_loc = '/content/'

#Query to load scoring data
training_data_query = client.query(config['training_data_query'])
test_data_query = client.query(config['test_data_query'])
test_data_query_ae = client.query(config['test_data_query_ae'])

#The independent and dependent variables which will be used in training and scoring
X = config['X']
y = config['y']
vars_outlier_treated = config['vars_outlier_treated']
vars_tobe_standardized = config['vars_tobe_standardized']

#cols and categories to be considered while outputing data to gbq
gbq_cols = config['gbq_cols']
top25_categories = config['top25_categories']



#----HELPER FUNCTIONS---#

#Module 1
def import_data_gbq(query, loc=None, filename=None):
  '''
  Helper function to import data from GBQ. 
  Based on the input query, the data is queried and output is saved in loc with the given filename.
  '''
  df = query.result().to_dataframe()
  if (loc != None) and (filename != None):
    df.to_pickle( str(loc) + str(filename) + '.pkl')
  return df


def clip_nth_percentile(df,cols,n_perc):
  '''
  Function to clip nth percentile in given columns
  '''
  for col in cols:
    df[col] = np.where(df[col]>df[col].quantile(n_perc),df[col].quantile(n_perc),df[col])
  return df

def brand_cat_clustering(df,n_clus=4,training=1,cluster_mapping=None):
  '''
  Clustering the brands and categories based on RPI, CTR and ATCPC based on KMeans clustering
  '''
  if training==1:
    #Aggregating the data at brand and category level for cliustering
    brand_clustering_df = df#[(df['brand_group_tailed_cleaned'] != 'brand_generic')&(df['category_derived'] != 'category_generic')].copy(deep=True)
    brand_clustering_df = brand_clustering_df.groupby(['brand_group_ol_cleaned','category_derived'],as_index=False).aggregate({'atc_updated':'sum','imp':'sum','clk':'sum','revenue':'sum','sku_config':'count'})
    brand_clustering_df['rpi'] = np.where(brand_clustering_df['imp']==0,0,(brand_clustering_df['revenue']/brand_clustering_df['imp']))
    brand_clustering_df['ctr'] = np.where(brand_clustering_df['imp']==0,0,(brand_clustering_df['clk']/brand_clustering_df['imp'])*100)
    brand_clustering_df['atcpc'] = np.where(brand_clustering_df['clk']==0, 0, (brand_clustering_df['atc_updated']/brand_clustering_df['clk']))
    scaler = StandardScaler()

    #Standardizing the cluster variables
    brand_clustering_df['rpi_standardized'] = scaler.fit_transform(np.array(brand_clustering_df['rpi']).reshape(-1, 1))
    brand_clustering_df['ctr_standardized'] = scaler.fit_transform(np.array(brand_clustering_df['ctr']).reshape(-1, 1))
    brand_clustering_df['atcpc_standardized'] = scaler.fit_transform(np.array(brand_clustering_df['atcpc']).reshape(-1, 1))
    brand_clustering_df = clip_nth_percentile(brand_clustering_df,cols=['rpi_standardized','ctr_standardized','atcpc_standardized'],n_perc=0.95)
    
    #Clustering input
    X = np.array(brand_clustering_df[['rpi_standardized','ctr_standardized','atcpc_standardized']])
    #Clustering
    kmeans = KMeans(n_clusters=n_clus)
    kmeans.fit(X)
    y_kmeans = kmeans.predict(X)
    brand_clustering_df['k_means_cluster'] =  y_kmeans
    brand_clustering_df['brand_category_cluster_label'] = brand_clustering_df['k_means_cluster'].apply(lambda x: 'cluster_' + str(int(x)))
    brand_cluster_mapping = brand_clustering_df[['brand_group_ol_cleaned','category_derived','brand_category_cluster_label']]
    #Joining back with the oriningal data
    df = df.merge(brand_cluster_mapping,on=['brand_group_ol_cleaned','category_derived'],how='left')
    return df,brand_cluster_mapping
  
  else:

    df = df.merge(cluster_mapping,on=['brand_group_ol_cleaned','category_derived'],how='left')
    return df


def fillna_cluster_cols(df):
  '''
  Wherever the cluster or imputed values are null, give them defaults
  '''
  df['brand_category_cluster_label'] = df['brand_category_cluster_label'].fillna('cluster_1')
  df['new_color'] = df['new_color'].fillna('event')
  df['new_occasion'] = df['new_occasion'].fillna('multicolor')
  return df

def standardize_data(df):
  for col in vars_outlier_treated:
    df[col] = np.where(df[col]>df[col].quantile(0.95),df[col].quantile(0.95),df[col])

  scaler = StandardScaler()
  df[vars_tobe_standardized] = scaler.fit_transform(df[vars_tobe_standardized])

  return df


def calc_rpi(combinations,names,df):
  '''
  Calculates RPI at any given level in the given dataframe
  '''

  for level, name in zip(combinations,names):
    temp = df.groupby(level,as_index=False).aggregate({'revenue':'sum','imp':'sum'})
    temp['rpi'] = np.where(temp['imp']==0, 0, (temp['revenue']/temp['imp']))
    temp['rpi'] = np.where(temp['rpi']>temp['rpi'].quantile(0.95),temp['rpi'].quantile(0.95),temp['rpi'])
    temp.rename({'rpi':name},axis=1,inplace=True)
    temp = temp[level + [name]]
    df = df.merge(temp,on=level,how='left')
  return df

def calc_atc_per_product(combinations,names,df):
  '''
  Calculates ATCPC at any given level in the given dataframe
  '''

  for level, name in zip(combinations,names):
    temp = df.groupby(level,as_index=False).aggregate({'atc':'sum','sku_config':'count'})
    temp['atc_pp'] = temp['atc']/temp['sku_config']
    temp['atc_pp'] = np.where(temp['atc_pp']>temp['atc_pp'].quantile(0.95),temp['atc_pp'].quantile(0.95),temp['atc_pp'])
    temp.rename({'atc_pp':name},axis=1,inplace=True)
    temp = temp[level + [name]]
    df = df.merge(temp,on=level,how='left')
  return df

def calc_mean_price(combinations,names,df):
  
  '''
  Calculates Mean price at any given level in the given dataframe
  '''
  for level, name in zip(combinations,names):
    temp = df.groupby(level,as_index=False).aggregate({'current_price':'mean'})
    temp.rename({'current_price':name},axis=1,inplace=True)
    temp = temp[level + [name]]
    df = df.merge(temp,on=level,how='left')
  return df

def training_add_vars_foldwise(df_input, prod_date = '2023-01-16'):
  '''
  Calculating the combination variables fold by fold

  '''

  df_non_zero = df_input[(df_input['rpi']!=0)&(pd.to_datetime(df_input['prod_date'])==pd.to_datetime(prod_date)) & ~((df_input['ageing']<30) & (df_input['imp']<410)) ]
  df_zero = df_input[(df_input['rpi']==0)&(pd.to_datetime(df_input['prod_date'])==pd.to_datetime(prod_date)) & ~((df_input['ageing']<30) & (df_input['imp']<410))]
  df_new = df_input[(pd.to_datetime(df_input['prod_date'])==pd.to_datetime(prod_date)) & ((df_input['ageing']<30) & (df_input['imp']<410))]
  n_sample_0 = (len(df_non_zero)+len(df_new)) //9

  #sampling the zeros based on thier impressions
  df_zero_sampled = df_zero.sort_values('imp',ascending=False)[:n_sample_0]

  #joining the zero sampled df with rest of the data
  df = pd.concat([df_non_zero,df_zero_sampled,df_new],axis=0)

  
  #We will create a combination of 2 variables from below list to calculate rpi
  levels_2_combinations =  ['brand_category_cluster_label','new_color','new_occasion',
                            'ageing_bucket','discount_bucket','price_bucket','season_grouped_ty']
                
  list_2_combinations=[]
  for item in combinations(levels_2_combinations, 2): 
    list_2_combinations.append(list(item))

  #We will create a combination of 1 variables from below list to calculate rpi
  levels_1_combinations =  ['brand_category_cluster_label','season_grouped_ty',
                              'new_color', 'ageing_bucket', 'price_bucket',
                            'discount_bucket']
                
  list_1_combinations=[]
  for item in combinations(levels_1_combinations, 1): 
    list_1_combinations.append(list(item))

  names_1_combinations = [('-'.join(i))+'_lp_rpi' for i in list_1_combinations]
  names_2_combinations = [('-'.join(i))+'_lp_rpi' for i in list_2_combinations]
  df = calc_rpi(list_2_combinations,names_2_combinations,df)
  df = calc_rpi(list_1_combinations,names_1_combinations,df)


  names_1_combinations = [('-'.join(i))+'_per_product_atc' for i in list_1_combinations]
  names_2_combinations = [('-'.join(i))+'_per_product_atc' for i in list_2_combinations]
  df = calc_atc_per_product(list_2_combinations,names_2_combinations,df)
  df = calc_atc_per_product(list_1_combinations,names_1_combinations,df)

  names_1_combinations = [('-'.join(i))+'_mean_price' for i in list_1_combinations]
  names_2_combinations = [('-'.join(i))+'_mean_price' for i in list_2_combinations]
  df = calc_mean_price(list_2_combinations,names_2_combinations,df)
  df = calc_mean_price(list_1_combinations,names_1_combinations,df)
  
  return df

def scoring_add_vars_foldwise(df):

  '''
  Calculating combination variables for scoring
  '''

  
  #We will create a combination of 2 variables from below list to calculate rpi
  levels_2_combinations =  ['brand_category_cluster_label','new_color','new_occasion',
                            'ageing_bucket','discount_bucket','price_bucket','season_grouped_ty']
                
  list_2_combinations=[]
  for item in combinations(levels_2_combinations, 2): 
    list_2_combinations.append(list(item))

  #We will create a combination of 1 variables from below list to calculate rpi
  levels_1_combinations =  ['brand_category_cluster_label','season_grouped_ty',
                              'new_color', 'ageing_bucket', 'price_bucket',
                            'discount_bucket']
                
  list_1_combinations=[]
  for item in combinations(levels_1_combinations, 1): 
    list_1_combinations.append(list(item))

  names_1_combinations = [('-'.join(i))+'_lp_rpi' for i in list_1_combinations]
  names_2_combinations = [('-'.join(i))+'_lp_rpi' for i in list_2_combinations]
  df = calc_rpi(list_2_combinations,names_2_combinations,df)
  df = calc_rpi(list_1_combinations,names_1_combinations,df)


  names_1_combinations = [('-'.join(i))+'_per_product_atc' for i in list_1_combinations]
  names_2_combinations = [('-'.join(i))+'_per_product_atc' for i in list_2_combinations]
  df = calc_atc_per_product(list_2_combinations,names_2_combinations,df)
  df = calc_atc_per_product(list_1_combinations,names_1_combinations,df)

  names_1_combinations = [('-'.join(i))+'_mean_price' for i in list_1_combinations]
  names_2_combinations = [('-'.join(i))+'_mean_price' for i in list_2_combinations]
  df = calc_mean_price(list_2_combinations,names_2_combinations,df)
  df = calc_mean_price(list_1_combinations,names_1_combinations,df)
  
  return df


def prepare_training_data(df,prod_dates_to_be_considered):
  '''
  Adding the all the independent variables to create training data
  '''
  df_training_fold = pd.DataFrame()
  #Adding the historical information to all the prod_dates
  for date in prod_dates_to_be_considered:
    print(date)
    if len(df_training_fold)>0:
      df_training_fold = pd.concat([df_training_fold,training_add_vars_foldwise(df,prod_date=date)],axis=0)
    else:
      df_training_fold = training_add_vars_foldwise(df,prod_date=date)
  return df_training_fold


def prepare_scoring_data(df):
  '''
  Adding the all the independent variables to create scoring
  '''
  df_scoring_fold = pd.DataFrame()
  df_scoring_fold = scoring_add_vars_foldwise(df)
  return df_scoring_fold


#Defining functions to calculate training error and validation error

def test_predict_and_error_estimate(model,xt,yt,df):
    '''
    Predict on the test/scoring data and estimate the error using ndcg score and MAE for top 25 categories
    '''
    prediction = model.predict(xt.reset_index(drop=True))
    xt.reset_index(drop=True,inplace=True)
    df.reset_index(drop=True,inplace=True)
    yt.reset_index(drop=True,inplace=True)
    #joining the model input with relevant data for the output
    scored_df = xt.merge(df[['sku_config','prod_date','brand_group_clean','brand_category_cluster_label','category_derived','category_clean','imp','ageing_bucket','discount_bucket',
                             'clk','revenue','atc','atc_updated','rpi','adjusted_rpi_score','atc_normalized','current_price','discount_pct']],left_index=True, right_index=True,how='left')

    scored_df = scored_df.merge(yt, left_index=True, right_index=True,how='left')
    #Adding predicted array to dataframe
    scored_df = pd.concat([scored_df.reset_index(drop=True),pd.Series(prediction)],axis=1)
    scored_df.rename(columns={ scored_df.columns[len(scored_df.columns)-1]: "Prediction_val" }, inplace = True)
    top_25_gdc = scored_df[scored_df['category_clean'].isin(top25_categories)].groupby(['category_clean'],as_index=False).agg({'brand_group_clean':'count','revenue':'sum'}).sort_values('brand_group_clean',ascending=False)
    top_25_gdc.reset_index(drop=True,inplace=True)
    top_25_gdc.rename({'brand_group_clean':'total_skus','revenue':'revenue_combination'},axis=1,inplace=True)
    
    #calculating ncdg at different values of k
    for k in [None,25,100,500]:
      top_25_gdc['ndcg_'+str(k)] = 0
      for index, row in top_25_gdc.iterrows():
        true_relevance = np.asarray([scored_df[(scored_df['category_clean']==row['category_clean'])]['relevance_updated']])
        predicted_relevance = np.asarray([scored_df[(scored_df['category_clean']==row['category_clean'])]['Prediction_val']])
        top_25_gdc.loc[index,'ndcg_'+str(k)] = np.round(ndcg_score(true_relevance, predicted_relevance, k=k),2)
        if k==500:
          top_25_gdc.loc[index,'mae'] = np.round(mean_absolute_error(true_relevance, predicted_relevance),2)
    return top_25_gdc,scored_df


def training(df_training):

  '''
  Function to train the model based on df_training
  Returns trained model

  '''

  #Making all the accidental nas zero
  df_training[X] = df_training[X].fillna(0)
  df_training[y] = df_training[y].fillna(0)

  #Takes around 7 mins to train
  df_training[X] = df_training[X].astype(float)
  df_training[y] = df_training[y].astype(float)

  xtrain = df_training[X]
  ytrain = df_training[y]
  ytrain['relevance_updated'] = ytrain['relevance_updated'].astype(int)

  xgbr = xgb.XGBRegressor(booster='gbtree',
  tree_method='gpu_hist',                       
  objective='reg:squarederror',
  random_state=42, 
  learning_rate=0.05,
  max_depth=4,
  min_child_weight=100, 
  n_estimators=500)

  xgbr.fit(xtrain[X],ytrain[y])

  return xgbr

def prediction(df_test,model):
  df_test[X] = df_test[X].fillna(0)
  df_test[y] = df_test[y].fillna(0)

  df_test[X] = df_test[X].astype(float)
  df_test[y] = df_test[y].astype(float)

  X_t = df_test[X]
  y_t = df_test[y]

  top_25_gdc,test_predicted_df =  test_predict_and_error_estimate(model,X_t,y_t,df_test)

  return test_predicted_df,top_25_gdc


def scoring(df_scoring,model):
  df_scoring[X] = df_scoring[X].fillna(0)
  df_scoring[y] = df_scoring[y].fillna(0)

  df_scoring[X] = df_scoring[X].astype(float)
  df_scoring[y] = df_scoring[y].astype(float)

  X_t = df_scoring[X]
  y_t = df_scoring[y]

  top_25_gdc,scored_df =  test_predict_and_error_estimate(model,X_t,y_t,df_scoring)

  return scored_df,top_25_gdc


def plot_SHAP(df_training,model):
  f = X.copy()
  feature_names= f
  df_training[X] = df_training[X].astype(float)
  shap.initjs()
  training_shap_input = df_training.sample(200000)
  explainer = shap.TreeExplainer(model)
  shap_values = explainer.shap_values(training_shap_input[X])
  shap.summary_plot(shap_values, features=training_shap_input[X], feature_names=feature_names,show=False)
  plt.savefig('feature_imp.png') 
  rf_resultX = pd.DataFrame(shap_values, columns = feature_names)
  vals = np.abs(rf_resultX.values).mean(0)
  shap_importance = pd.DataFrame(list(zip(feature_names, vals)),
                                    columns=['col_name','feature_importance_vals'])
  shap_importance.sort_values(by=['feature_importance_vals'],
                                ascending=False, inplace=True)
  return shap_importance

def training_data_snapshot(df_training,tr_version):
  '''
  Function to snapshot the training data so that drift can be calcated
  '''
  df_training['train_version'] = tr_version
  df_training['date'] = df_training['train_version'].apply(lambda x : x.split('.')[0])
  df_training['batch'] = df_training['train_version'].apply(lambda x : x.split('.')[1])
  df_training['run_type'] = df_training['train_version'].apply(lambda x : x.split('.')[2])
  df_training['train_version_number'] = df_training['train_version'].apply(lambda x : int(re.sub('[^0-9]+', '', x)))
  op_cols = gbq_cols + ['train_version','date','batch','run_type','train_version_number','brand_group_ol_cleaned','category_ol_cleaned','color_ol_cleaned','occasion_ol_cleaned']
  df_training[op_cols].to_gbq("namshi-datagrill.ai_ranking.training_data_snapshot",
                      project_id='namshi-datagrill',
                      chunksize=10000,
                      reauth=False,
                      if_exists='append',
                      auth_local_webserver=True,
                      location='US',
                      progress_bar=True,
                      credentials=None)


def output_brand_cat_cluster(df_cluster,version):
  '''
  Output categorywise NDCG, MAE details
  '''
  df_cluster['version'] = version
  df_cluster.to_gbq("namshi-datagrill.ai_ranking.dim_brand_cat_cluster",
                    project_id='namshi-datagrill',
                    chunksize=10,
                    reauth=False,
                    if_exists='replace',
                    auth_local_webserver=True,
                    location='US',
                    progress_bar=True,
                    credentials=None)
  
def add_version_number(df,version_score,version_train,country='SA'):
  '''
  Function to add version number and country to output tables
  '''
  df['version'] = version_score
  df['train_version'] = version_train
  df['country'] = country
  df['date'] = df['version'].apply(lambda x : x.split('.')[0])
  df['batch'] = df['version'].apply(lambda x : x.split('.')[1])
  df['run_type'] = df['version'].apply(lambda x : x.split('.')[2])
  df['test_version_number'] = df['version'].apply(lambda x : int(re.sub('[^0-9]+', '', x)))
  df['train_version_number'] = df['train_version'].apply(lambda x : int(re.sub('[^0-9]+', '', x)))
  return df

def add_bucket_variables_op(df):
  '''
  Function to add custom ageing bucket and rank bucket for the output
  '''
  df['ageing_bucket_new'] = np.select([(df['ageing']<=30) & (df['ageing']>0) ,
                                                (df['ageing']<=60) & (df['ageing']>30),
                                                (df['ageing']<=90) & (df['ageing']>60),
                                                (df['ageing']<=180) & (df['ageing']>90),
                                                (df['ageing']<=360) & (df['ageing']>180),
                                                (df['ageing']<=720) & (df['ageing']>360),
                                                (df['ageing']>720)],
                                              ['1: 0-30',
                                                '2: 31-60',
                                                '3: 61-90',
                                                '4: 91-180',
                                                '5: 181-360',
                                                '6: 361-720',
                                                '7: 720+'],'Misc')

  df['rank_bucket'] = np.select([(df['category_clean_rank']<=10) & (df['category_clean_rank']>0) ,
                                                (df['category_clean_rank']<=20) & (df['category_clean_rank']>10),
                                                (df['category_clean_rank']<=30) & (df['category_clean_rank']>20),
                                                (df['category_clean_rank']<=40) & (df['category_clean_rank']>30),
                                                (df['category_clean_rank']<=50) & (df['category_clean_rank']>40),
                                                (df['category_clean_rank']<=60) & (df['category_clean_rank']>50),
                                                (df['category_clean_rank']<=70) & (df['category_clean_rank']>60),
                                                (df['category_clean_rank']<=80) & (df['category_clean_rank']>70),
                                                (df['category_clean_rank']<=90) & (df['category_clean_rank']>80),
                                                (df['category_clean_rank']<=100) & (df['category_clean_rank']>90),
                                                (df['category_clean_rank']<=300) & (df['category_clean_rank']>100),
                                                (df['category_clean_rank']<=1000) & (df['category_clean_rank']>300),
                                                (df['category_clean_rank']>1000)],
                                              ['1: 0-10',
                                                '2: 11-20',
                                                '3: 21-30',
                                                '4: 31-40',
                                                '5: 41-50',
                                                '6: 51-60',
                                                '7: 61-70',
                                                '8: 71-80',
                                                '9: 81-90',
                                                '10: 91-100',
                                                '11: 100-300','12: 300-1000','13: 1000+'],'Misc')
  return df


def output_GBQ_scoring(scored_df,df_scoring_copy,df_error,cols,topcats,time_taken,version_score,version_train,test=1,country='SA'):
  '''
  The predictions from the model at sku_config level, error details, time and version information are consolidated and appended to \
  "namshi-datagrill.ai_ranking.output_prediction_snapshot" table in GBQ
  '''

  #Processing scored_df
  scored_df.columns = [x.replace('-','_') for x in scored_df.columns]
  scored_df = scored_df.sort_values('Prediction_val',ascending=True)
  scored_df.reset_index(drop=True,inplace=True)
  scored_df['predicted_relevance'] = np.round(scored_df['Prediction_val'])
  scored_df['predicted_relevance']  = np.where(scored_df['predicted_relevance']>10,10,scored_df['predicted_relevance'])
  scored_df['predicted_relevance']  = np.where(scored_df['predicted_relevance']<1,1,scored_df['predicted_relevance'])

  #Processing df_scoring_copy
  df_scoring_copy.columns = [x.replace('-','_') for x in df_scoring_copy.columns]
  
  #Creating df_gbq_scoring_op by mergeing scored_df and df_scoring_copy
  df_gbq_scoring_op = scored_df[['sku_config','prod_date','Prediction_val','predicted_relevance']].merge(df_scoring_copy[cols], on=['sku_config','prod_date'])
  df_gbq_scoring_op.rename({'Prediction_val':'prediction'},axis=1,inplace=1)
  df_gbq_scoring_op.sort_values('prediction',ascending=False,inplace=True)
  df_gbq_scoring_op = df_gbq_scoring_op[df_gbq_scoring_op['availability_ksa_pct']>0]
  df_gbq_scoring_op['category_clean_rank'] = df_gbq_scoring_op.groupby('category_clean',as_index=False)['prediction'].rank(ascending=False,method='first')
  df_gbq_scoring_op['ageing'] = df_gbq_scoring_op['ageing'].astype(float)
  df_gbq_scoring_op = add_bucket_variables_op(df_gbq_scoring_op)
  df_gbq_scoring_op[['category_clean_rank','prod_date']] = df_gbq_scoring_op[['category_clean_rank','prod_date']].astype(str)
  df_gbq_scoring_op[['co_flag','noos_flag','core_flag','Crossdocking_flag','relaunch_flag']] = df_gbq_scoring_op[['co_flag','noos_flag','core_flag','Crossdocking_flag','relaunch_flag']].astype(int) 
  #Adding error details from df_error 
  df_gbq_scoring_op = df_gbq_scoring_op.merge(df_error[['category_clean', 'total_skus', 'revenue_combination', 'ndcg_None','ndcg_25', 'ndcg_100', 'ndcg_500', 'mae']],on='category_clean',how='left')
  #Adding runtime information for test
  df_gbq_scoring_op['time_taken'] = time_taken
  #Adding version details
  df_gbq_scoring_op = add_version_number(df_gbq_scoring_op,version_score,version_train,country=country)

  df_gbq_scoring_op['ageing_less_than30'] = np.where(df_gbq_scoring_op['ageing']<30, 1,0)
  df_gbq_scoring_op['FP'] = np.where(df_gbq_scoring_op['discount_pct']<=0, 1,0)

  df_gbq_scoring_op.to_gbq("namshi-datagrill.ai_ranking.output_prediction_snapshot",
                    project_id='namshi-datagrill',
                    chunksize=10000,
                    reauth=False,
                    if_exists='append',
                    auth_local_webserver=True,
                    location='US',
                    progress_bar=True,
                    credentials=None)     
  
def output_GBQ_scoring_current_algo(df_scoring_copy,cols,topcats,version_score,version_train,test=1,country='SA'):

  #Creating scored_df
  scored_df = pd.read_json("https://storage.googleapis.com/pages-scores-live/exp_2_scores.json")
  if country=='AE':
    scored_df = pd.DataFrame(scored_df['ae']).reset_index()
  else:
    scored_df = pd.DataFrame(scored_df['sa']).reset_index()

  scored_df.columns = ['sku_config','prediction']
  scored_df = scored_df.sort_values('prediction',ascending=True)

  #Processing df_scoring_copy
  df_scoring_copy.columns = [x.replace('-','_') for x in df_scoring_copy.columns]

  #Merging scored_df and df_scoring_copy
  df_gbq_scoring_op = scored_df[['sku_config','prediction']].merge(df_scoring_copy[cols], on=['sku_config'])
  df_gbq_scoring_op.sort_values('prediction',ascending=False,inplace=True)
  df_gbq_scoring_op = df_gbq_scoring_op[df_gbq_scoring_op['availability_ksa_pct']>0]
  df_gbq_scoring_op['category_clean_rank'] = df_gbq_scoring_op.groupby('category_clean',as_index=False)['prediction'].rank(ascending=False,method='first')
  df_gbq_scoring_op['ageing'] = df_gbq_scoring_op['ageing'].astype(float)
  df_gbq_scoring_op = add_bucket_variables_op(df_gbq_scoring_op)
  df_gbq_scoring_op[['category_clean_rank','prod_date']] = df_gbq_scoring_op[['category_clean_rank','prod_date']].astype(str)
  df_gbq_scoring_op[['co_flag','noos_flag','core_flag','Crossdocking_flag','relaunch_flag']] = df_gbq_scoring_op[['co_flag','noos_flag','core_flag','Crossdocking_flag','relaunch_flag']].astype(int) 
  df_gbq_scoring_op = add_version_number(df_gbq_scoring_op,version_score,version_train,country=country)
  df_gbq_scoring_op['ageing_less_than30'] = np.where(df_gbq_scoring_op['ageing']<30, 1,0)
  df_gbq_scoring_op['FP'] = np.where(df_gbq_scoring_op['discount_pct']<=0, 1,0)

  df_gbq_scoring_op.to_gbq("namshi-datagrill.ai_ranking.output_prediction_current_algo_snapshot",
                    project_id='namshi-datagrill',
                    chunksize=10000,
                    reauth=False,
                    if_exists='append',
                    auth_local_webserver=True,
                    location='US',
                    progress_bar=True,
                    credentials=None)
     

def plot_confusion_matrix(scored_df):
  '''
  plotting confusion matrix

  '''
  temp = scored_df.sort_values('Prediction_val',ascending=True)
  temp.reset_index(drop=True,inplace=True)
  temp['predicted_relevance'] = np.round(temp['Prediction_val']) #(np.ceil(temp.index//(len(temp)//10)))+1
  temp['predicted_relevance']  = np.where(temp['predicted_relevance']>10,10,temp['predicted_relevance'])
  temp['predicted_relevance']  = np.where(temp['predicted_relevance']<1,1,temp['predicted_relevance'])
  temp2 = temp.groupby(['relevance_updated','predicted_relevance'],as_index=False)['sku_config'].count()
  #sns.scatterplot(data=scored_df,x='relevance',y='pred_interval')
  plt.figure(figsize=(20,10))
  sns.scatterplot(data=temp2,x='relevance_updated',y='predicted_relevance',size='sku_config',sizes=(20,1000))
  plt.show()
  return temp2,temp  
```
